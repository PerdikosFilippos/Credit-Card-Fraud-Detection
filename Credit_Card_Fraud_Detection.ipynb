#Import the libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#Data Analysis

```python
#loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

```python
#print first 5 rows
credit_card_data.head()
```

```python
#information on dataset
credit_card_data.info()
```

```python
#check the number of mising values in each column
credit_card_data.isnull().sum()
```

```python
#check distribution of legit transactions and faudulent transactions
credit_card_data["Class"].value_counts()
```

**Very unbalanced Dataset**

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measurements of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()

```

```python
#Compare the values for both legit and fraud transactions
credit_card_data.groupby("Class").mean()
```

**UnderSampling**

#Build a sample dataset containing similar distribution between legit and fraud transactions. ie: Make the legit ones 492, same as fraud

```python
legit_sample = legit.sample(n=492)
```

#Concancenate the two dataframes into a new dataset

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
#check the two datasets
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

#Split data into features and targets

```python
X = new_dataset.drop(columns="Class", axis=1)
Y = new_dataset["Class"]
```

#Split data into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Train model using Logistic Regression

```python
model = LogisticRegression()
```

```python
#training logistic regression model with training data
model.fit(X_train, Y_train)
```

#Evaluate model using accuracy score

```python
#Evaluate accuracy on TRAINING data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on Training Data: ", training_data_accuracy)
```

```python
#Evaluate accuracy on TEST data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on Test Data: ", test_data_accuracy)
```
